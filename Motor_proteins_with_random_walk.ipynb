### 1a)
The diffusion equation is given by the expression
$$
\frac{\partial \Phi(x, t)}{\partial t} = D \frac{\partial^2 \Phi(x, t)}{\partial x^2} \tag{1.1}.
$$
And a proposed solution with
$$
\phi(x, t) = \frac{1}{\sqrt{4\pi Dt}} \int_{-\infty}^\infty h(y) e^{-\frac{(x-y)^2}{4Dt}} dy \tag{1.2}.
$$
We want to show that $(1.2)$ is a solution to the diffusion equation. We start by taking the fourier transform of $(1.1)$ on both sides with respect to $x$. Let $\widehat\Phi(k,t) = \mathcal{F} \left\{ \Phi(x,t) \right\}$. Using properties for the fourier transform, we find that

$$
\mathcal{F} \left\{ \frac{\partial \Phi}{\partial t} \right\}
= \frac{\partial \widehat \Phi}{\partial t},
\quad
\mathcal{F} \left\{ \frac{\partial^2 \Phi}{\partial x^2} \right\}
= -k^2 \widehat\Phi(k,t).
$$
Which gives us the diffusion equation in $k$ - space. Which is as first order ODE in terms of t.
$$
\frac{\partial \widehat \Phi}{\partial t} = -Dk^2 \widehat\Phi(k,t) \tag{1.3}
$$

The solution of which is given by the expression

$$
\widehat\Phi(k, t) = \widehat\Phi(k, 0) e^{-D k^2 t} \tag{1.4}.
$$

Since this is a product, the simplest way to take the inverse fourier transform is by using the convolution theorem. Which is given by the equation below

$$
\frac{1}{\sqrt{2\pi}} (f * g)(t) = \mathcal{F^{-1}}\left\{ \widehat{f(t)} \cdot \widehat{g(t)} \right\} \tag{1.5}
$$

We can find the inverse of $e^{-Dk^2t}$ by looking it up in a table. The left handside of $(1.4)$ then becomes

\begin{equation}
\begin{aligned}
\phi(x,t) = \mathcal{F}^{-1}\left\{ \widehat\Phi(k, 0) e^{-D k^2 t} \right\} &= \frac{1}{\sqrt{2\pi}} \Bigl( \widehat\Phi(x,0) * \frac{1}{\sqrt{2Dt}} e^{-\frac{x^2}{4Dt}} \Bigr) \\
&= \frac{1}{\sqrt{4\pi Dt}} \int_{-\infty}^{\infty} \widehat\Phi(y,0)\,e^{-\frac{(x-y)^2}{4Dt}}\,dy.
\end{aligned}
\tag{1.6}
\end{equation}

This equation $(1.6)$ matches the proposed solution $(1.2)$ if $\widehat\Phi(y,0)=h(y)$. This means that $h(y)$ is the inital distribution of particles along the x-axis.

### 1b)


The probability of the particle being in position $x$ is given by:
$$P(x) = \frac{e^{-\beta V(x)}}{Z} \tag{1.7}$$


$Z$ is the partition function, defined as:

$$Z = \sum_{i} e^{-\beta V(x_i)}.  \tag{1.8}$$

where V is a potential function and $\beta$ = $\frac{1}{k_bT}$, where $k_b$ is boltzmanns constant and $T$ is the temperature. $x_i$ is the particle position.

Since the particle is only able to perform movement to the left, right or standing still.
The probabilities for moving to specific positions are then given as $p^+ = \frac{e^{-\beta V(x_0 + 1)}}{Z}$, which is the probability of moving to the right, in the same manner the probability of moving to the left is $p^- = \frac{e^{-\beta V(x_0 - 1)}}{Z}$ and $p^0 = \frac{e^{-\beta V(x_0)}}{Z}$ is the probability of the paricle standing still.

To get the desired expressions given in the exercise, we manipulate $p^+$,$p^-$ and $p^0$:

$$p^+ = \frac{e^{-\beta V(x_0+1)}}{e^{-\beta V(x_0+1)} + e^{-\beta V(x_0-1)} + e^{-\beta V(x_0)}} \cdot \frac{e^{\beta V(x_0+1)}}{e^{\beta V(x_0+1)}} = \frac{1}{1 + e^{-\beta [V(x_0-1) - V(x_0+1)]} + e^{-\beta [V(x_0) - V(x_0+1)]}}\ \tag{1.9}$$

$$p^- = \frac{e^{-\beta V(x_0+1)}}{e^{-\beta V(x_0-1)} + e^{-\beta V(x_0-1)} + e^{-\beta V(x_0)}} \cdot \frac{e^{\beta V(x_0-1)}}{e^{\beta V(x_0-1)}} = \frac{1}{1 + e^{-\beta[V(x_0+1) - V(x_0-1)]} + e^{-\beta[V(x_0) - V(x_0-1)]}} \tag{1.10}
$$

$$p^0 = \frac{e^{-\beta V(x_0)}}{e^{-\beta V(x_0-1)} + e^{-\beta V(x_0-1)} + e^{-\beta V(x_0)}} \cdot \frac{e^{\beta V(x_0)}}{e^{\beta V(x_0)}} = \frac{1}{1 + e^{-\beta[V(x_0-1) - V(x_0)]} + e^{-\beta[V(x_0+1) - V(x_0)]}}\tag{1.11}
$$

### 1c)

We start by looking at the first case when $k_b T \gg |(V(x+1) - V(x))|$. We know that $\beta = \frac{1}{k_b T}$, so a large $k_b T$ will make $\beta \to 0$.
Looking at the expressions above (1.9), (1.10), and (1.11), we take the limit when $\beta$ goes to zero:
$$ p^+ = p^0 = p^- = \frac{1}{3} $$ 
This shows that when the thermal energy scale, represented by $k_BT$, is much larger than the potential energy scale, there is an equal probability of moving in any direction. In this limit, the random walk can be simplified by assuming that all movements are equally probable.

---

For the second case, the situation is different. When $k_b T \ll |(V(x+1) - V(x))|$, we have $\beta \to \infty$. We start by looking at $p^+$:

$$
p^+ = \frac{1}{1 + e^{-\beta [V(x_0-1) - V(x_0+1)]} + e^{-\beta [V(x_0) - V(x_0+1)]}}.
$$

If we consider the case where $V(x_0+1) < V(x_0-1)$ and $V(x_0+1) < V(x_0)$, then both $e^{-\beta [V(x_0-1) - V(x_0+1)]}$ and $e^{-\beta [V(x_0) - V(x_0+1)]}$ will have negative exponents. Taking the limit of this expression as $\beta \to \infty$, the terms with the negative exponents go to 0, leaving:

$$
p^+ = 1.
$$

For this case, $p^0$ and $p^-$ will both be 0

---

Next, we look at $p^0$:

$$
p^0 = \frac{1}{1 + e^{-\beta [V(x_0-1) - V(x_0)]} + e^{-\beta [V(x_0+1) - V(x_0)]}}.
$$

If we consider the case where $V(x_0) < V(x_0-1)$ and $V(x_0) < V(x_0+1)$, then both $e^{-\beta [V(x_0-1) - V(x_0)]}$ and $e^{-\beta [V(x_0+1) - V(x_0)]}$ will have negative exponents. Taking the limit of this expression as $\beta \to \infty$, the terms with the negative exponents go to 0, leaving:

$$
p^0 = 1.
$$

In this case, $p^+$ and $p^-$ will both be 0.

---

Finally, we look at $p^-$:

$$
p^- = \frac{1}{1 + e^{-\beta [V(x_0+1) - V(x_0-1)]} + e^{-\beta [V(x_0) - V(x_0-1)]}}.
$$

If we consider the case where $V(x_0-1) < V(x_0)$ and $V(x_0-1) < V(x_0+1)$, then both $e^{-\beta [V(x_0+1) - V(x_0-1)]}$ and $e^{-\beta [V(x_0) - V(x_0-1)]}$ will have negative exponents. Taking the limit of this expression as $\beta \to \infty$, the terms with the negative exponents go to 0, leaving:

$$
p^- = 1.
$$

For this case, $p^+$ and $p^0$ will both be 0.

So when $k_b T \ll |(V(x+1) - V(x))|$, the probabilities depend on the smallest potential:

If $V(x_0+1)$ is the smallest, then:
  $$
  p^+ = 1, \quad p^0 = 0, \quad p^- = 0.
  $$
If $V(x_0)$ is the smallest, then:
  $$
  p^+ = 0, \quad p^0 = 1, \quad p^- = 0.
  $$
and finally, if $V(x_0-1)$ is the smallest, then:
  $$
  p^+ = 0, \quad p^0 = 0, \quad p^- = 1.
  $$

So the simplification that can be made for the random walk model in this scenario is that the movement will be completely determined by the potential, which will move the particle in the direction which lowers the potential energy, this is in agreement with the fact that the force is $F = -\nabla V$

We start by importing neccesary libraries and defining the given parameters for the simulation

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from scipy import special
```

```python
n_particles = 10_000
delta_t = 1
h = 1
steps = 200
beta_k_values = [0.01, 1, 100]
bin_min = -steps
bin_max = steps
num_bins = 100
bins = np.linspace(bin_min, bin_max, num_bins)
```

Then we create a mutual simulation function to simulate the random walk for the diffrent potentials given in exercise 2




```python
def simulate_particles(potential_func, Bk = beta_k_values , steps = steps, num_particles = n_particles, h=h,initial_particles = None):
    """
    Simulates particle movement under a given potential function with discrete time steps.

    Parameters
    ----------
    potential_func : callable
        Function that calculates movement probabilities for particles.
        Returns tuple (p_right, p_left) probabilities for right/left moves
    Bk : array-like, optional
        Array of beta values to simulate.
        Default: beta_k_values
    steps : int, optional
        Number of discrete time steps in simulation. Default: steps
    num_particles : int, optional
        Number of independent particles to simulate. Default: n_particles
    h : int, optional
        Step size for particle displacement. Default: h
    initial_particles: array-like, optional.
        Initial particle distribution
    Returns
    -------
    numpy.ndarray
        Array of shape (num_particles, len(Bk)) containing final positions:
        - Columns correspond to different beta values in Bk.
        - Rows represent individual particles positions.
    """
    # Initialize particle positions
    if initial_particles is None:
      particles = np.zeros((num_particles, len(Bk)))
    else:
      particles = initial_particles

    # Loop over each beta value
    for i, bk in enumerate(Bk):
        # Loop over each step
        for j in range(steps):

            # Generate random numbers for all particles
            rand = np.random.rand(num_particles)

            # Get probabilities from the potential function for moving right and left
            prob = potential_func(particles[:, i], bk)
            p_pos, p_neg = prob[0], prob[1]

            # Update particle positions based on random numbers and probabilities
            mov_left = (rand <= p_neg)  # Particles moving left
            mov_right = (rand > (1 - p_pos))  # Particles moving right
            particles[mov_left, i] -= h  # Move left
            particles[mov_right, i] += h  # Move right
    return particles
```

Substituting the constant potential, the expression becomes:
$$
p^+ = \frac{1}{1 + e^{-\beta[k - k]} + e^{-\beta[k -k]}} = \frac{1}{1 + e^{-\beta k[1-1]} + e^{-\beta k[1 -1]}} = \frac{1}{3}
$$

As we can see the $\beta k$ dependency cancels out and the probability is a constant.

in the same manner it can be shown that $p^- = \frac{1}{3}$ and $p^0 = \frac{1}{3}$.

Therefore it is evident that the random walk simulation for $\beta k \in \{0.01, 1, 100\}$ will give the same results, with any differences only because of stochastic variations. Because of this we are only plotting it once.




```python
def Const_pot(x, bk):
    """
    Constant potential

    Returns
    -------
    tuple (p_plus, p_minus)
        p_plus: Probability to move right
        p_minus: Probability to move left    
    """
    p_right = 1/3  # Right move probability
    p_left = 1/3  # Left move probability
    return p_right, p_left
```

From the theory section we have a way of connecting the random walk model to the diffusion equation.

Let $X(t)$ be the position of a particle at time $t$. The probability distribution of the particle's position is given by $\phi(x, t) = P\{X(t) = x\} .$

The evolution of $\phi(x, t)$ can be described by:

$$\phi(x, t + \Delta t) = p^+ \phi(x - h, t) + p^- \phi(x + h, t) + p^0 \phi(x, t)$$
$$= \frac{1}{3} \phi(x - h, t) + \frac{1}{3} \phi(x + h, t) + \frac{1}{3} \phi(x, t)\tag{2.1}$$

Substracting $\phi(x, t)$ from both sides and dividing by $\Delta t$ we end up with:

$$\frac{\phi(x, t + \Delta t) - \phi(x, t)}{\Delta t} = \frac{h^2}{3\Delta t} \frac{\phi(x - h, t) - 2\phi(x, t) + \phi(x + h, t)}{h^2}$$

If we define $ D = \frac{h^2}{3\Delta t}$ as the diffusion constant and let $ \Delta t$ and $h$ approach zero, the right-hand side becomes the second derivative of $\phi$ with respect to $x$ and the left-hand side becomes the partial derivative of $\phi$ with respect to time

$$\frac{\partial \phi(x, t)}{\partial t} = D \frac{\partial^2 \phi(x, t)}{\partial x^2}$$

From the exercise we have that all particles will be at $x=0$ when $t=0$, therefore the intitial intial distribution will be $\delta(y)$ .
$$
\phi(x, t) = \frac{1}{\sqrt{4\pi D t}} \int_{-\infty}^{\infty}\,\delta(y) e^{-\frac{(x - y)^2}{4Dt}}\mathrm{d}y  = \frac{1}{\sqrt{4\pi D t}} e^{-\frac{x^2}{4Dt}}\tag{2.2}
$$

As we can see the solution is a normal curve with $\mu=0$ and $\sigma=\sqrt{2Dt}$ Since we have simulated 200 steps where the time for each step is 1s, $t=200$
.So $\sigma$ of our fitted normal-curve should be $\sqrt{\frac{2}{3}\cdot200} \approx 11.54$ and $\mu ≈ 0$. Looking at the values given by the normal fit, they seem to match well.

### 2b)


We start by inserting the given potential $V(x) = -kx$ for $p^+, p^-,$ and $p^0$ to find the $\beta k$ dependencies. Using the given potential, we end up with the following expressions for the probabilities:

$$
p^+ = \frac{1}{1 + e^{-\beta[-k(x_0-1)+k(x_0+1)]} + e^{-\beta[-kx_0+k(x_0+1)]}} = \frac{1}{1 + e^{-2\beta k} + e^{-\beta k}}\tag{2.3}
$$

$$
p^- = \frac{1}{1 + e^{-\beta[-k(x_0+1)+k(x_0-1)]} + e^{-\beta[-kx_0+k(x_0-1)]}} = \frac{1}{1 + e^{2\beta k} + e^{\beta k}}\tag{2.4}
$$

$$
p^0 = \frac{1}{1 + e^{-\beta[-k(x_0-1)+kx_0]} + e^{-\beta[-k(x_0+1)+kx_0]}} = \frac{1}{1 + e^{\beta k} + e^{-\beta k}}\tag{2.5}
$$


```python
def probabilities_linear_potential(x,bk):
    """
    Compute probabilities (right, left) for a particle in a linear potential V(x) = -kx.

    Parameters
    ----------
    beta_k : beta_k_values
        Ratio of potential energy scale to thermal energy (βk = k/(k_B T)).

    Returns
    -------
    tuple (p_plus, p_minus)
        p_plus: Probability to move right
        p_minus: Probability to move left
    """
    p_plus = 1/(1+np.exp(-2*bk)+ np.exp(-bk))
    p_minus = 1/(1+np.exp(2*bk)+ np.exp(bk))
    return p_plus,p_minus
```

### 2c)

```python
def trig_potential_function(x):
    """
    Compute the potential V(x) = k(x/15 - cos(x/3)).

    Parameters
    ----------
    x : numpy.ndarray or float
        Position(s) where the potential is evaluated.

    Returns
    -------
    float or numpy.ndarray
        Potential value(s) at the given position(s).
    """
    return x / 15 - np.cos(x / 3)

def potential_probabilities(x, beta_k):
    """
    Compute probabilities (right, left) for a particle based on the potential function.
    Parameters
    ----------
    x : numpy.ndarray
        Current positions of particles.
    beta_k : float
        Ratio of potential energy scale to thermal energy (βk = k/(k_B T)).

    Returns
    -------
    tuple
        p_plus: Probability to move right
        p_minus: Probability to move left
    """
    # Compute potential values
    v_x = trig_potential_function(x)
    v_x_plus_1 = trig_potential_function(x + 1)
    v_x_minus_1 = trig_potential_function(x - 1)

    # Compute probabilities using given expressions
    p_plus = 1 / (1 + np.exp(-beta_k * (v_x_minus_1 - v_x_plus_1)) + np.exp(-beta_k * (v_x - v_x_plus_1)))
    p_minus = 1 / (1 + np.exp(-beta_k * (v_x_plus_1 - v_x_minus_1)) + np.exp(-beta_k * (v_x - v_x_minus_1)))

    return p_plus, p_minus
```

```python
final_positions = simulate_particles(potential_probabilities)
fig, axes = plt.subplots(1, 3, figsize=(17, 5))
fig.suptitle(r'Particle Distribution for $V(x) = k\left(\frac{x}{15} - \cos \frac{x}{3}\right)$', fontsize=16)

for i, (ax, beta_k) in enumerate(zip(axes, beta_k_values)):
    if beta_k == 100:
        bins = 4

    else:
        bins = 'scott'

    n, bins, patches = ax.hist(final_positions[:, i], bins=bins, density=True,alpha=1, label="Simulated distribution")

    mu, sigma = scipy.stats.norm.fit(final_positions[:, i])

    x_fit = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

    ax.plot(x_fit, scipy.stats.norm.pdf(x_fit, mu, sigma), 'k-', linewidth=2, label=f'Fit: μ={mu:.2f}\nσ={sigma:.2f}')

    ax.set_xlim(mu - 4*sigma, mu + 4*sigma)

    ax.set_xlabel('Position')
    ax.set_ylabel('Density')
    ax.set_title(rf'$\beta k = {beta_k}$')
    ax.legend()
    ax.grid()
    ax.axvline(0, color='r', linestyle='--', linewidth=1)

    label = f"fig: 1.{i+1}"
    ax.text(0.05, 0.95, label, transform=ax.transAxes,fontsize=14, va='top')

plt.tight_layout()
plt.show()
```

When we run the simulation with $\beta k = 0.01$, the thermal energy dominates the weak potential. In practice, we observe that the particles perform nearly unbiased random walks, so their positions spread out widely.

When $\beta k = 1$, the thermal energy and potential energy scales become comparable. The particles are influenced by the potential and they will also diffuse randomly. Because the potential will drag the particles to the closest minima. To locate these minima, we solve $V'(x) = 0$

Our potential function does not have any global minima, but the function values get smaller, as we decrease x. Within the range $-25 < x < 0$, the solution for $V'(x) = 0$  is approximately $-0.603$ and $-19$, so theese are the closest minima of the function in the range.

Particles tend to cluster around these two minima. Most particles are trapped near the first minimum at $x \approx -0.603$, becauase its closest, while some escape because of random diffusion and get trapped at the second minimum at $x \approx -19.452$. This explains why figure 1.2 shows a distribution with two peaks: a larger peak near $x \approx -0.603$ and a smaller peak near $x \approx -19.452$.

### 2d)


We start by deriving the movement probabilities for a particle moving under the influence of the quartic potential:

$$
V(x) = kx^4\tag{2.6}
$$

By substituting $V(x) = kx^4$ into equations $(1.9)$, $(1.10)$ og $(1.11)$, we obtain:

$$
p^+ = \frac{1}{1 + e^{\beta k \left( 8x^3 + 8x \right)} + e^{\beta k \left( 4x^3 + 6x^2 + 4x + 1 \right)}}
$$

$$
p^- = \frac{1}{1 + e^{-\beta k \left( 8x^3 + 8x \right)} + e^{-\beta k \left( 4x^3 - 6x^2 + 4x - 1 \right)}}
$$


```python
def quartic_pot(x, bk):
    """
    Compute probabilities (right, left, stay) for a particle in a quartic potential.

    Parameters
    ----------
    beta_k : list
        Ratio of potential energy scale to thermal energy (βk = k/(k_B T)).

    Returns
    -------
    tuple (p_plus, p_minus)
        p_plus: Probability to move right
        p_minus: Probability to move left
    """
    # Probabilities for moving left or right
    p_pos = 1/(1+np.exp(bk*(8*x**3+8*x))+np.exp(bk*(4*x**3+6*x**2+4*x+1)))
    p_neg = 1/(1+np.exp(-bk*(8*x**3+8*x))+np.exp(-bk*(4*x**3-6*x**2+4*x-1)))
    return p_pos, p_neg
```

```python
def find_potential_region(positions, alpha, N_x, n_sawtooth_peaks):
    """
    Determine which piece of the potential V_1(x) the given x-position belongs to,
    with the x-axis discretized from 0 to (n_sawtooth_peaks * N_x - 1).

    Parameters:
    positions : float or np.ndarray
        Current (local) positions of each particle (shape = (n_particles,)).
    alpha : float
        Sawtooth parameter in (0,1).
    N_x : int
        Period length of the potential V_1(x).
    n_sawtooth_peaks : int
        Number of sawtooth peaks.

    Returns:
    np.ndarray
        Array indicating region 1 or 2 for each position.
    """
    total_period = n_sawtooth_peaks * N_x

    # Ensure all positions are within valid range (correctly wrap around)
    positions = np.where(positions >= total_period, positions % total_period, positions)

    # Compute modular position within a single period
    positions_mod = positions % N_x

    # Define region boundaries
    region1_start = 0
    region1_end = alpha * N_x
    region2_start = region1_end
    region2_end = N_x

    # Assign regions
    regions = np.zeros_like(positions_mod, dtype=int)
    mask1 = (positions_mod > region1_start) & (positions_mod <= region1_end)
    mask2 = (positions_mod > region2_start) & (positions_mod <= region2_end)

    regions[mask1] = 1  # Assign region 1
    regions[mask2] = 2  # Assign region 2

    return regions
```

```python
def calculate_probabilities_V1(positions, alpha, N_x, b_k, n_sawtooth_peaks):
    """
    Calculate the probabilities p_plus and p_minus for V1(x),
    We first find which region each position is in, then assign
    different  (p_plus, p_minus) for region 1 and region 2.

    Parameters
    ----------
    positions : np.ndarray
        Current (local) positions of each particle (shape = (n_particles,)).
    alpha : float
        Sawtooth parameter in (0,1).
    N_x : int
        Period length of the potential V_1(x).
    b_k : float
        Ratio of potential energy scale to thermal energy (βk = k/(k_B T)).

    n_sawtooth_peaks : int
        Number of sawtooth peaks.

    Returns
    -------
    p_plus, p_minus : np.ndarray
        Arrays of shape (n_particles,) giving the probability
        that each particle moves to right or left.
    """
    region = find_potential_region(positions, alpha, N_x, n_sawtooth_peaks)

    p_plus = np.zeros_like(positions, dtype=float)
    p_minus = np.zeros_like(positions, dtype=float)

    #To suppress floating-point errors such as division by zero, overflow and underflow
    with np.errstate(over='ignore', under='ignore'):

        # Region 1
        factor_r1 = b_k / (alpha*N_x)
        
        exp2_r1 = np.exp(2 * factor_r1)
        exp1_r1 = np.exp(factor_r1)
        p_plus_r1  = 1.0 / (1.0 + exp2_r1 + exp1_r1)

        exp_neg2_r1 = np.exp(-2 * factor_r1)
        exp_neg1_r1 = np.exp(-factor_r1)
        p_minus_r1 = 1.0 / (1.0 + exp_neg2_r1 + exp_neg1_r1)

        # Region 2 
        factor_r2 = b_k / ((1 - alpha)*N_x)

        exp_neg2_r2 = np.exp(-2*factor_r2)
        exp_neg1_r2 = np.exp(-factor_r2)
        p_plus_r2  = 1.0 / (1.0 + exp_neg2_r2 + exp_neg1_r2)

        exp_pos2_r2 = np.exp(2*factor_r2)
        exp_pos1_r2 = np.exp(factor_r2)
        p_minus_r2 = 1.0 / (1.0 + exp_pos2_r2 + exp_pos1_r2)

    mask1 = region == 1
    mask2 = region == 2

    p_plus[mask1]  = p_plus_r1
    p_minus[mask1] = p_minus_r1
    p_plus[mask2]  = p_plus_r2
    p_minus[mask2] = p_minus_r2

    return p_plus, p_minus

```

```python
def probabilities_V2(x, bk):
    """
    Compute the probabilities for right and left moves in a constant potential.
    Parameters:
    ----------
    x : int
        The position of the particle. This parameter is not used in the computation 
        since the potential is constant.
    bk : float
        Ratio of potential energy scale to thermal energy.
    Returns:
    -------
    tuple
        A tuple `(p_plus, p_minus)` representing the probabilities of moving right.
    """
    p_plus = 1 / 3  
    p_minus = 1 / 3  
    return p_plus, p_minus
```

```python
def is_in_V2_phase(t,T_p):
  """
  Check if the particle is supposed to be in V_1 or V_2 for the current timestep

  Returns
  -------
    bool
        True if the system is in the V2 phase at the current time step, 
        False if it is in the V1 phase.

  """
  if (t % (2*T_p))<T_p:
    return True
  else:
    return False
```

```python
def simulate_random_walk_modified(n_cycles,alpha, T_p, N_x, b_k, n_particles, h,n_sawtooth_peaks,):
    """
    Runs the random-walk simulation
    Parameters
    ----------
    alpha : float
        Sawtooth parameter in (0,1).
    T_p : int
        Number of steps to spend in each phase.
    N_x : int
        period length of the potential V_1(x).
    b_k : float
        Ratio of potential energy scale to thermal energy (βk = k/(k_B T)).
    n_particles : int
        Number of particles
    h : float
        Step length

    Returns
    -------
    positions_cycle_end : np.ndarray, shape (n_cycles, n_particles)
        The absolute positions of each particle at the end of each cycle.
    """
    
    total_steps = (2 * T_p) * n_cycles

    # local positions and absolute positions
    x_local = np.zeros(n_particles)
    X_abs   = np.zeros(n_particles)

    # store (n_cycles, n_particles) absolute positions at cycle ends
    positions_cycle_end = np.zeros((n_cycles, n_particles), dtype=int)
    cycle_count = 0

    for t in range(total_steps):
        if is_in_V2_phase(t,T_p):
            p_plus = np.full(n_particles,probabilities_V2(x_local,b_k)[0])
            p_minus = np.full(n_particles, probabilities_V2(x_local,b_k)[1])
        else:
            p_plus, p_minus = calculate_probabilities_V1(x_local,alpha,N_x,b_k,n_sawtooth_peaks)

        # Generating random values for particles to determine their movement and updating positions
        rand = np.random.rand(n_particles)
        move_left  = rand <= p_minus
        move_right = rand > 1.0 - p_plus

        # Update local and absolute positions based on movements
        x_local = (x_local - (move_left*h) + (move_right*h)) % (n_sawtooth_peaks*N_x)
        X_abs = X_abs - (move_left*h) + (move_right*h)

        # Storing positions at the end of each cycle
        if (t+1) % (2*T_p) == 0: # Check if the end of a cycle is reached
            positions_cycle_end[cycle_count, :] = X_abs
            cycle_count +=1

    return positions_cycle_end
```

```python
T_p = 200
alpha = 0.1
b_k = 1000
N_p = 3
N_x = 100

def simulate_random_walk_modified(n_cycles, alpha, T_p, N_x, b_k, n_particles, h, n_sawtooth_peaks, part_per_index):
    """
    Simulates a random walk for particles under a periodic potential.
    
    Parameters
    ----------
    n_cycles : int
        Number of cycles for the simulation. Each cycle consists of 2*T_p steps.
    alpha : float
        Sawtooth parameter in (0,1), influencing the asymmetry of the V1 potential.
    T_p : int
        Number of steps to spend in each phase (V1 and V2).
    N_x : int
        Period length of each sawtooth peak in the V1 potential.
    b_k : float
        Ratio of the potential energy scale to thermal energy.
    n_particles : int
        Total number of particles in the simulation.
    h : int
        Step length of particles during each movement.
    n_sawtooth_peaks : int
        Total number of sawtooth peaks in the periodic potential.
    part_per_index : int
        Number of particles to initialize at each spatial index within the system.

    Returns
    -------
    positions_cycle_end : np.ndarray, shape (n_cycles, n_particles)
        The absolute positions of each particle at the end of each cycle.
    J : np.ndarray, shape (total_steps,)
        The particles current for each timestep

    """

    total_steps = (2 * T_p) * n_cycles

    # local positions and absolute positions
    x_local = np.repeat(np.arange(0, 2 * N_x), part_per_index)
    X_abs = np.copy(x_local)

    left_moves = np.zeros(total_steps, dtype=int)
    right_moves = np.zeros(total_steps, dtype=int)
    J = np.zeros(total_steps)

    # store (n_cycles, n_particles) absolute positions at cycle ends
    positions_cycle_end = np.zeros((n_cycles, n_particles), dtype=int)
    cycle_count = 0
    for t in range(total_steps):
        # Determine if in phase V2 or V1 and calculating phase-specific or sawtooth respectively
        if is_in_V2_phase(t, T_p):
            p_plus = np.full(n_particles, probabilities_V2(x_local, b_k)[0])
            p_minus = np.full(n_particles, probabilities_V2(x_local, b_k)[1])
        else:
            p_plus, p_minus = calculate_probabilities_V1(x_local, alpha, N_x, b_k, n_sawtooth_peaks)

        # Randomized particle movement direction and decide left or right movement based on probabilities
        rand = np.random.rand(n_particles)
        move_left = rand <= p_minus
        move_right = rand > 1.0 - p_plus

        # Counting left/right moving particles
        left_moves[t] = np.sum(move_left)
        right_moves[t] = np.sum(move_right)

        # Calculating and updating positions of particles
        J[t] = (right_moves[t]-left_moves[t])/n_particles
        x_local = (x_local - (move_left * h) + (move_right * h)) % (n_sawtooth_peaks * N_x)
        X_abs = X_abs - (move_left * h) + (move_right * h)

        # Storing absolute positions of all particles
        if (t + 1) % (2 * T_p) == 0:
            positions_cycle_end[cycle_count, :] = X_abs
            cycle_count += 1

    return positions_cycle_end, J
```

```python
def simulate_one_cycle(alpha, T_p, N_x, b_k, n_particles, h, n_sawtooth_peaks):
    """
    Simulates a random walk for particles under a periodic potential.

    Parameters
    ----------
    alpha : float
        Sawtooth parameter in (0,1), influencing the asymmetry of the V1 potential.
    T_p : int
        Number of steps to spend in each phase (V1 and V2).
    N_x : int
        Period length of each sawtooth peak in the V1 potential.
    b_k : float
        Ratio of the potential energy scale to thermal energy (βk = k/(k_B T)).
    n_particles : int
        Total number of particles in the simulation.
    h : int
        Step length of particles during each movement.    
    n_sawtooth_peaks : int
        Total number of sawtooth peaks in the periodic potential.

    Returns
    -------
    J_avg : float
        The average net current over the full cycle.
    """
   
    # Total time steps for one cycle (combining both V1 and V2 phases)
    total_steps = 2 * T_p

    # Initialize particle positions:
    # Half of the particles start at position 0, the other half at position N_x.
    x_local = np.concatenate([np.full(n_particles // 2,0), np.full(n_particles - n_particles // 2, 2* N_x)])

    # Variable to accumulate the net current over the cycle
    current_sum = 0.0
    current_J = np.zeros(total_steps)
    # Loop over each time step in the cycle
    for t in range(total_steps):
        # Determine phase based on the current time step.
        if is_in_V2_phase(t, T_p):
            # In V2 phase, use uniform probabilities for all particles.
            p_plus_val, p_minus_val = probabilities_V2(x_local, b_k)
            p_plus = np.full(n_particles, p_plus_val)
            p_minus = np.full(n_particles, p_minus_val)
        else:
            # In V1 phase, probabilities are computed based on individual positions.
            p_plus, p_minus = calculate_probabilities_V1(x_local, alpha, N_x, b_k, n_sawtooth_peaks)

        # Generate a random number for each particle to decide its movement.
        rand = np.random.rand(n_particles)

        # Determine which particles move left or right:
        # If the random number is <= p_minus, the particle moves left.
        move_left = (rand <= p_minus)
        # If the random number is > (1 - p_plus), the particle moves right.
        move_right = (rand > (1 - p_plus))

        # Count the number of particles moving right and left.
        n_plus = np.sum(move_right)
        n_minus = np.sum(move_left)

        # Calculate the net current at this time step.
        current_J[t] = (n_plus - n_minus) / n_particles

        # Update particle positions:
        # Particles moving left subtract h from their position,
        # particles moving right add h, and positions wrap around using modulo.
        x_local = (x_local - (move_left * h) + (move_right * h)) % (n_sawtooth_peaks * N_x)

    # Compute and return the average current over the full cycle.
    J_avg = np.sum(current_J) / total_steps
    return J_avg
```

For small values of $ T_p $, the net current approaches zero because the rapid switching between $ V_1(x) $ and $ V_2(x) $ prevents the particles from being significantly influenced in either direction, leading to no directional movement.
```python
def J_analytic(alpha, N_x, T_p):
    """
    Calculates the average particle current J_avg over one cycle
    in a ratchet potential system, based on the complementary error function (erfc).

    Parameters
    ----------
    alpha : float or ndarray
        Sawtooth parameter in (0,1), influencing the asymmetry of the V1 potential.
    N_x : int
        The spatial period of the sawtooth potential
    T_p : int
        Number of steps to spend in each phase (V1 and V2).
    Returns
    -------
    float or ndarray
        The cycle-averaged current J_avg.
    """
    exp1 = (alpha * N_x) / (2) * np.sqrt(3 / T_p)
    exp2 = ((1 - alpha) * N_x) / (2) * np.sqrt(3 / T_p)

    return (N_x / (4 * T_p)) * (special.erfc(exp1) - special.erfc(exp2))
```

```python
plt.plot(alpha_values,J_analytic(alpha_values,100,500),label = "Analytic")
plt.plot(alpha_values, J_avg_values, label='$J_{avg}$')
plt.xlabel("Alpha values")
plt.ylabel("$J_{avg}$")
plt.legend()
plt.grid(True)
plt.show()
```

We observe that the analytical and numerical values of the cycle-averaged current $ J_{\text{avg}} $ coincide very well for $ T_p = 500 $ and $\beta k = 1000$. Obviously the anlaytical solution will be smooth, becuase it is solved in a continious manner, while the numerical is discretized.  

As in 3d), the parameter $ \alpha $ controls the asymmetry between the regions, influencing $ J_{\text{avg}} $. $ \alpha > 0.5 $ increases the size of region 1, favoring leftward motion ($ n^- $), while $ \alpha <0.5 $ enlarges region 2, favoring rightward motion ($ n^+ $), which is also reflected in the given equations for the analytic solution. There are some more subtle points about the size of $T_p$ and $\beta k$ that are required for the analytical to coincide with the numerical, that we will discuss later.

### 4d)

The analytical solution assumes that by the end of each cycle, all particles have settled into energy minima. This is largely influenced by the cycle duration, $T_p$, which determines how much time particles have to diffuse before the next potential shift occurs.  However, if $T_p$ is too small, particles may not have sufficient time to reach a minimum, leading to deviations from the expected analytical results, as shown in task 4b). In extreme cases, of low $T_p$ rapid switching between the sawtooth potential and diffusion phases can cause particles to behave more randomly, reducing the influence of the potential and diminishing the net current.

Another key factor is the parameter $\beta k$, which dictates the ratio between the potetnial energy and thermal energy. When $\beta k$ is large, the potential energy dominates, and particles predominantly follow the structured sawtooth potential. In this phase, the analytical assumptions hold well, as particles are mostly guided into energy minima by the potential. However, for low $\beta k$ values, thermal energy begins to dominate, introducing a significant degree of randomness into the particle motion, as shown in task 4c). When thermal fluctuations outweigh the potential influence, particle movement increasingly resembles a random walk, making it less likely that they will end up in the predicted minima. This leads to discrepancies between the analytical predictions and numerical results.

The analytical solution also assumes that during each cycle, particles move at most one step in either direction before settling into the nearest energy minimum. This assumption is implicitly made in the derivation of n⁺ and n⁻, where the probability of movement is determined by integrating the diffusion distribution up to a single threshold $\alpha N_x$ or $−(1−\alpha)N_x$. This means the model does not consider the possibility of particles traveling beyond the closest well in a single cycle.

For very large $T_p$ or small $\beta k$, diffusion can become significant enough that some particles travel multiple steps before being recaptured in a potential well. When thermal energy is comparable to or greater than the potential energy barrier, particle motion starts to resemble a random walk rather than deterministic movement toward the nearest minimum. This effect is especially pronounced at low $\beta k$, where particles have enough energy to escape multiple wells before being trapped.
